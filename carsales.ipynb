```python
car['Customer Name'] = car['Customer Name'].fillna('unknown')
```

```python
car.isnull().sum()
```

```python
car["Full name"] = car['Company'] + " " + car['Model']
```

```python
car['Gender'].value_counts() 
```

```python
car['Gender']=car['Gender'].map({'Male':0,'Female':1})
```

```python
car['Body Style'].value_counts() 
```

```python
car['Transmission'].value_counts()
```

```python
car['Transmission']=car['Transmission'].map({'Auto':0,'Manual':1})
```

```python
car['Engine'].value_counts()
```

```python
car['Engine'] = car['Engine'].str.strip()
```

```python
car['Engine'] = car['Engine'].str.replace('Â','', regex=False)
car['Engine'] = car['Engine'].str.replace('\xa0',' ', regex=False)
```

```python
car['Engine'] = car['Engine'].map({'Double Overhead Camshaft': 0,'Overhead Camshaft': 1})
```

```python
car['Engine'].value_counts()
```

```python
car['Color'].value_counts() 
```

```python
car['Date']=pd.to_datetime(car['Date'],errors='coerce')
```

```python
car['Month'] = car['Date'].dt.month
car['Year']  = car['Date'].dt.year
```

```python
def season_from_month(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'
car['Season'] = car['Month'].apply(season_from_month)
```

```python
car.drop(['Dealer_No ', 'Car_id','Phone'], axis=1, inplace=True)
```

```python
car.groupby('Gender')['Price ($)'].mean().plot(kind='bar',color=['blue','red'])
plt.title('Average Car Price by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Price ($)')
plt.show()                            
```

```python
color_counts = car.groupby(['Gender','Color']).size().unstack(fill_value=0)
palette = {'Pale White': 'gray', 'Red': 'red', 'Black': 'black'}
color_counts.plot(kind='bar', stacked=False, figsize=(10,6), color=palette)
plt.title('Car Color Preference by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Cars')
plt.show()
```

```python
sns.histplot(car['Price ($)'],color='skyblue',bins=20,edgecolor='black')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Number of Cars')
plt.show()
```

```python
sns.countplot(car['Dealer_Region'],color='skyblue')
```

```python
max_price = car['Price ($)'].max()

most_expensive_car = car[car['Price ($)'] == max_price]

print(most_expensive_car.T)
```

```python
min_price = car['Price ($)'].min()
cheapest_car = car[car['Price ($)'] == min_price]
print(cheapest_car.T)
```

```python
seasonal_price = car.groupby('Season')['Price ($)'].max().sort_values(ascending=False)
print(seasonal_price)
```

```python
seasonal_price.plot(kind='bar',color='orange')
plt.title('Most Expensive Cars by Season')
plt.xlabel('Season')
plt.ylabel('($)')
plt.show()
```

```python
monthly_sales_revenue = car.groupby('Month')['Price ($)'].agg(['sum', 'mean'])
print(monthly_sales_revenue)
```

```python
season_stats = car.groupby('Season')['Price ($)'].agg(['min', 'max', 'mean']).round(2)
print(season_stats)
```

```python
cheapest_cars = car.loc[car.groupby('Season')['Price ($)'].idxmin()]
most_expensive_cars = car.loc[car.groupby('Season')['Price ($)'].idxmax()]
print(cheapest_cars.T)
```

```python
numeric_cols = car.select_dtypes(include='number')
correlation_matrix = numeric_cols.corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numeric Columns')
plt.show()
```

```python
sns.boxplot(car['Annual Income'])
```

```python
Q1 = car['Annual Income'].quantile(0.25)
Q3 = car['Annual Income'].quantile(0.75)
IQR = Q3 - Q1
# Choosing outliers
annual_outliers = car[(car['Annual Income'] < Q1 - 1.5*IQR) | 
                      (car['Annual Income'] > Q3 + 1.5*IQR)]

print("Number of outliers in Annual Income:", len(annual_outliers))
print(annual_outliers[['Customer Name','Annual Income','Price ($)']])
```

```python
sns.boxplot(car['Price ($)'])
```

```python
Q1 = car['Price ($)'].quantile(0.25)
Q3 = car['Price ($)'].quantile(0.75)
IQR = Q3 - Q1
# Choosing outliers
price_outliers = car[(car['Price ($)'] < Q1 - 1.5*IQR) | (car['Price ($)'] > Q3 + 1.5*IQR)]

print("Number of outliers in Annual Income:", len(annual_outliers))
print(price_outliers[['Customer Name','Annual Income','Price ($)']])
```
